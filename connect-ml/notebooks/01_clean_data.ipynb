```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

print("✅ All imports working")
```

```python
jobs = pd.read_csv("../data/raw/postings.csv")

print(f"Total rows: {jobs.shape[0]}")
print(f"Total columns: {jobs.shape[1]}")
print("\nColumn names:")
print(jobs.columns.tolist())
```

```python
jobs.head(3)
```

```python
print("Missing values per column:")
print(jobs.isnull().sum())
```

```python
# Only these 3 columns matter for Career Path Predictor
jobs_clean = jobs[['title', 'description', 'formatted_experience_level']].copy()

# Drop rows where title or description is missing
jobs_clean = jobs_clean.dropna(subset=['title', 'description'])

# Lowercase and strip whitespace from job titles
jobs_clean['title'] = jobs_clean['title'].str.lower().str.strip()

print(f"✅ Clean dataset: {jobs_clean.shape[0]} rows")
jobs_clean.head(3)
```

```python
CAREER_DOMAINS = {
    "Software Engineering": [
        "software engineer", "backend engineer", "frontend engineer",
        "full stack developer", "software developer", "sde"
    ],
    "Data & AI": [
        "data scientist", "machine learning engineer", "data analyst",
        "ai engineer", "data engineer", "mlops engineer"
    ],
    "Product Management": [
        "product manager", "associate product manager", "senior product manager",
        "apm", "growth product manager"
    ],
    "Design": [
        "ux designer", "ui designer", "product designer",
        "interaction designer", "visual designer"
    ],
    "Finance": [
        "investment banker", "financial analyst", "equity research",
        "business analyst", "management consultant"
    ],
    "Marketing": [
        "marketing manager", "digital marketing", "content strategist",
        "brand manager", "growth hacker"
    ]
}

def tag_domain(title):
    title = str(title).lower()
    for domain, keywords in CAREER_DOMAINS.items():
        if any(keyword in title for keyword in keywords):
            return domain
    return "Other"

jobs_clean['domain'] = jobs_clean['title'].apply(tag_domain)

print("Domain distribution:")
print(jobs_clean['domain'].value_counts())
```

```python
jobs_final = jobs_clean[jobs_clean['domain'] != 'Other'].copy()

print(f"Final dataset size: {jobs_final.shape[0]} rows")

# Save
jobs_final.to_csv("../data/processed/jobs_with_domains.csv", index=False)
print("✅ Saved to data/processed/jobs_with_domains.csv")
```
